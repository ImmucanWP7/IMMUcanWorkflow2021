## Convert czi files to tiff

Here, we will first convert the acquired czi files to tiff files for processing.

```python
import os
import xtiff
from aicsimageio import AICSImage
import numpy as np
from xtiff import TiffProfile

for file in os.listdir("/Volumes/immucan_volume/raw_data/Panel_1/2022_WORKFLOW/czi/"):
    cur_file = AICSImage(os.path.join("/Volumes/immucan_volume/raw_data/Panel_1/2022_WORKFLOW/czi/", file))
    first_channel = cur_file.get_image_dask_data("ZYX", C=0, S=0, T=0)
    second_channel = cur_file.get_image_dask_data("ZYX", C=1, S=0, T=0)
    third_channel = cur_file.get_image_dask_data("ZYX", C=2, S=0, T=0)
    forth_channel = cur_file.get_image_dask_data("ZYX", C=3, S=0, T=0)
    all_channels = np.stack([first_channel, second_channel, third_channel, forth_channel], axis=0)

    xtiff.to_tiff(np.squeeze(all_channels), os.path.join("/Volumes/immucan_volume/processed_data/Panel_1/2022_WORKFLOW/mIF/raw", os.path.splitext(file)[0] + ".tiff"), profile=TiffProfile.TIFF)
```
